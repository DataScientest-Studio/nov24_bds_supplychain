```python
import pandas as pd
import emoji
from scipy.stats import chi2_contingency

# Charger le fichier CSV
df = pd.read_csv("C:\\Users\\amina\\df_projet_truspilot_2.csv")

# V√©rifiez si les colonnes "Note client" et "Commentaire" existent
if "Note client" in df.columns and "Commentaire" in df.columns:
    # Ajouter une colonne pour indiquer la pr√©sence d'√©motic√¥nes
    df["Pr√©sence_√âmotic√¥nes"] = df["Commentaire"].apply(
        lambda commentaire: 1 if isinstance(commentaire, str) and any(char in emoji.EMOJI_DATA for char in commentaire) else 0)
    

    df["Cat√©gorie_Note"] = pd.cut(
        df["Note client"],
        bins=[0, 2, 4, 5],  
        labels=["Faible", "Moyenne", "√âlev√©e"],
        include_lowest=True)

    table_contingence = pd.crosstab(df["Cat√©gorie_Note"], df["Pr√©sence_√âmotic√¥nes"])
    print("Table de contingence :")
    print(table_contingence)

    # Test du chi-carr√©
    resultats_chi2 = chi2_contingency(table_contingence)
    p_value = resultats_chi2[1]
    statistique = resultats_chi2[0]
    print("Test du khi2 d'ind√©pendance")
    print(" ")
    print("ùêª0 : La note du client est ind√©pendante de la pr√©sence d'√©motic√¥nes dans le commentaire")
    print("ùêª1 : La note du client n'est pas ind√©pendante de la pr√©sence d'√©motic√¥nes dans le commentaire")
    print(" ")
    print("La statistique du test est : ", statistique, "\n"
      "La p-valeur du test est : ", p_value, "\n")
    alpha=0.05

    if p_value < alpha:
        print("Nous rejetons l'hypoth√®se nulle H0.")
        print("La note du client d√©pend de la pr√©sence d'√©motic√¥nes dans le commentaire.")
    else:
        print("Nous ne rejetons pas l'hypoth√®se nulle H0.")
        print("La note du client ind√©pendante de la pr√©sence d'√©motic√¥nes dans le commentaire.")


#######################################################################################################################
#------------------------------------- R√©sultat -----------------------------------------------------------------------
####################################################################################################################"##
''''
Table de contingence :
Pr√©sence_√âmotic√¥nes      0    1
Cat√©gorie_Note                 
Faible                3142   47
Moyenne               5689   49
√âlev√©e               24284  392
Test du khi2 d'ind√©pendance
 
ùêª0 : La note du client est ind√©pendante de la pr√©sence d'√©motic√¥nes dans le commentaire
ùêª1 : La note du client n'est pas ind√©pendante de la pr√©sence d'√©motic√¥nes dans le commentaire
 
La statistique du test est :  17.566940524975568 
La p-valeur du test est :  0.00015324535976142298 

Nous rejetons l'hypoth√®se nulle H0.
La note du client d√©pend de la pr√©sence d'√©motic√¥nes dans le commentaire.


'''
```
