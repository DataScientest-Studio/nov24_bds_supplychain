```python
###############################################################################------PREPROCESSING-----####################################################################################################
#                                                                    
#                                                                                    NETTOYAGE
#                                                                                    EXPLORATION  de la base scrapée précédemment                                                                                   
#                                                                                    DATA VIZ 
#                                                              
##############################################################################################################################################################################################################

################################################## - RECUPERATION de la base scrapée projet csv - ##################################

import pandas as pd

df = pd.read_csv(r"C:\Users\magal\Documents\Documents Magali\Formations\Formation machine learing\DataScientist\projet\code\df_projet_truspilot2.csv", sep=',')
print(len(df))
display(df.head(5))

################################################ - IMPORTATION des bibliotheques - ##################################################

%matplotlib inline    
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import locale
from datetime import datetime
from scipy.stats import pearsonr

# Configuration la locale en français
locale.setlocale(locale.LC_TIME, 'fr_FR.UTF-8')

#Vérification des noms des mois dans la biliotheque calendar python
import calendar

mois_complets = list(calendar.month_name)[1:]  # Liste des noms de mois (1 à 12)
mois_abreges = list(calendar.month_abbr)[1:]    # Liste des noms abrégés (1 à 12)

print(mois_complets)  # Affiche ['Janvier', 'Février', 'Mars', ...]
print(mois_abreges)   # Affiche ['Jan', 'Fév', 'Mar', ...]


################################################ - NETTOYAGE - #################################################################
###############################################################################################################################

print("_________________________________________Info________________________________________________")

# Informations sur le DataFrame
print(df.info())

print("_________________________________________Format date _________________________________________")

# Traitement des formats date
from datetime import datetime as dt

#Remplacer les mois accentués par des équivalents sans accent
df['Date de publication'] = df['Date de publication'].replace({
    "février": "fÃ©vrier",
    "août": "aoÃ»t",
    "décembre": "dÃ©cembre"
}, regex=True)

df['Date de publication'] = pd.to_datetime(df['Date de publication'], format="%d %B %Y", dayfirst=True)


print("_____________________________________Valeurs manquantes_____________________________________")

# Vérification des valeurs manquantes
display("valeurs manquantes:\n", df.isnull().sum())

# Remplace les valeurs manquantes de la colonne Commentaire par le contenu de la colonne titre (personne n'ayant pas rempli)
df['Commentaire'] = df['Commentaire'].fillna(df['Titre commentaire'])


print("______________________________________40 doublons supprimés____________________________________")

# Vérification des valeurs manquantes
print("Nombre de doublons :", df.duplicated().sum())

#Supprime les doublons 
df= df.drop_duplicates() # enlever le # pour le 1er go

############################################## - EXPLORATION - #############################################################
#############################################################################################################################

print("__________________________________________ Stat __Note client __________________________________")

print (df['Note client'].describe())
print(" ")
print(df.boxplot('Note client'))

# Nombre total d'événements
total_avis = df.shape[0]

# Nombre de notes 1-5 dans la variable Note client
nbre_note1 = (df['Note client'] == 1).sum()
nbre_note2 = (df['Note client'] == 2).sum()
nbre_note3 = (df['Note client'] == 3).sum()
nbre_note4 = (df['Note client'] == 4).sum()
nbre_note5 = (df['Note client'] == 5).sum()

print(f"Nombre total avis : {total_avis}")

print(f"Nombre de note 1: {nbre_note1}")
print(f"Nombre de note 2: {nbre_note2}")
print(f"Nombre de note 3: {nbre_note3}")
print(f"Nombre de note 4: {nbre_note4}")
print(f"Nombre de note 5 : {nbre_note5}")


print("_______________________________Répartition des notes et entreprises________________________________")

# note
noteclient_counts = df['Note client'].value_counts()
print(" Repartition des notes clients :\n", noteclient_counts)
print(" ")

# entreprise 
entreprise_counts = df['Nom entreprise'].value_counts()
print(" Repartition du poids des entreprise:\n", entreprise_counts)
print(" ")

print("____________________________________Periode d'observation__________________________________________")

print(" ")
date_max = df['Date de publication'].max()
date_min = df['Date de publication'].min()

print(f"Période d'observation : {date_min.date()} à {date_max.date()}")
print(" ")

print("_______________________Correlation entre le nombre d'avis du client et la note ____________________")


# Calcul de la corrélation de Pearson
correlation, p_value = pearsonr(df['Note client'], df["Nombre d'avis du client"])

print(f'Coefficient de corrélation de Pearson : {correlation}')
print(f'Valeur p : {p_value}')

# Interprétation
if p_value < 0.05:
    print("Il y a une corrélation significative entre la note et le nombre d'avis laissés par le client.")
else:
    print("Il n'y a pas de corrélation significative entre la note et le nombre d'avis laissés par le client.")

 # Création du nuage de points
plt.figure(figsize=(8, 6))
plt.scatter(df['Note client'], df["Nombre d'avis du client"], color='blue', marker='o')
plt.title("Note du client vs Nombre d'avis")
plt.xlabel("Note du client")
plt.ylabel("Nombre d'avis")

# Affichage de la corrélation sur le graphique
plt.text(1, 100, f'Corrélation : {correlation:.2f}', fontsize=12, color='green')
plt.text(1, 95, f'Valeur p : {p_value}', fontsize=12, color='green')
plt.show()
print(" ")

#-------------------------------------------------------------------- SAUVEGARDE d'un nouveau DataFrame  --------------------------------------------------------------------------------------
#                                                           avec les retraitements qui servira de base de travail 
#
#df.to_csv("C:\\Users\\magal\\Documents\\Documents Magali\\Formations\\Formation machine learing\\DataScientist\\projet\\code\\df_projet_truspilot2_etape2.csv", index=False, encoding='utf-8-sig')

#-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------


############################################## - DATA VIZ - ###############################################################
###########################################################################################################################



print("_____________________1 Graph descriptif___Répartition des entreprises_________________________________")

materiel = (df['Nom entreprise'] == "Materiel.net").sum()
Rebuy = (df['Nom entreprise'] == "Rebuy").sum()
Recommerce = (df['Nom entreprise'] == "Recommerce.com").sum()

# Création des listes de labels et de tailles
labels = ['Materiel.net', 'Rebuy', 'Recommerce.com']
sizes = [materiel, Rebuy, Recommerce]
colors = ['#ff9999', '#66b3ff', '#99ff99']  # Couleurs pour chaque secteur
explode = (0.1, 0, 0)  # Mettre en avant Materiel.net


# Création du diagramme circulaire
plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  
plt.title('Répartition des entreprises')
plt.show()


print("_________________________________________2 Graph descriptifs_courbe par année____________________________________")

df['année'] = df['Date de publication'].dt.year
an = df['année'].value_counts().sort_index()

# Courbe nombre total d'avis par an
plt.figure(figsize=(10, 6))
sns.lineplot(x=an.index, y=an.values, marker='o')
plt.title("Évolution de nombre d'avis par année")
plt.xlabel("Année")
plt.ylabel("Nombre d'avis")
plt.xticks(rotation=45)  
plt.grid() 
plt.show()  

print("_______________________________Graph descriptifs_courbe par année et par note client_________________________________")


# Nombre d'avis par année et par note
publication_counts = df.groupby(['année', 'Note client']).size().reset_index(name='count')


plt.figure(figsize=(10, 6))
sns.lineplot(data=publication_counts, x='année', y='count', hue='Note client', marker='o',palette="Set1")
plt.title("Évolution nombre d'avis selon la note attribuée par année")
plt.xlabel("Année")
plt.ylabel("Nombre d'avis")
plt.xticks(rotation=45)  
plt.grid()  
plt.legend(title='Note client')  
plt.show() 


print("_______________________4 Graph descriptif___Répartition des notes clients par entreprise_________________________________")


# Calcule la fréquence des notes par entreprise
count_df = df.groupby(['Nom entreprise', 'Note client']).size().unstack(fill_value=0)

# Calcule les pourcentages
percentage_df = count_df.div(count_df.sum(axis=1), axis=0) * 100

# Création de la figure
plt.figure(figsize=(10, 6))

# Ajouter un fond gris clair
plt.gca().set_facecolor('#f0f0f0')

# Création de l'histogramme empilé horizontal
percentage_df.plot(kind='barh', stacked=True, colormap='tab10', ax=plt.gca())
plt.title('Répartition des notes clients par entreprise')
plt.xlabel('Pourcentage des notes clients')
plt.ylabel('Nom entreprise')
plt.legend(title='Note client', bbox_to_anchor=(1.05, 1), loc='upper left')

# Affiche les pourcentages à côté des barres
for i in range(len(percentage_df)):
    for j in range(len(percentage_df.columns)):
        plt.text(percentage_df.iloc[i, :j+1].sum() - percentage_df.iloc[i, j] / 2,
                 i, f'{percentage_df.iloc[i, j]:.1f}%', ha='center', va='center', color='white')

# Supprime le contour du graphique
sns.despine()

plt.tight_layout()
plt.show()


print("_______________________3 Graph descriptif___Répartition des notes clients par entreprise___________________________________")

# noteclient_counts = df['Note client'].value_counts()
# print(" Repartition des notes clients :\n", noteclient_counts)

# Création de la figure
plt.figure(figsize=(10, 6))

# Création du countplot
ax = sns.countplot(data=df, x='Note client', palette='viridis')
plt.title('Répartition des notes clients')
plt.xlabel('Note client')
plt.ylabel('Nombre de clients')
plt.xticks(rotation=0)

# Supprime le contour du graphique
sns.despine()

# Déplace la légende à l'extérieur
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Affiche le nombre au-dessus des barres
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='bottom')
plt.tight_layout()  
plt.show()
```
